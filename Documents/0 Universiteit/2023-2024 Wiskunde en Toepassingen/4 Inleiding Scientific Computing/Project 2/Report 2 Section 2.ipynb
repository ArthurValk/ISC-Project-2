```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.draw import rectangle
from skimage.transform import radon, iradon
from skimage.data import shepp_logan_phantom
```

```python
nx = 1024 # number of pixels and angles used to show sinogram

theta = np.linspace(0., 180., nx)

# create uniform square on [-1,1]^2
ii, jj = rectangle(start=(nx//4, nx//4), end=(3*nx//4, 3*nx//4))
u = np.zeros((nx, nx))
u[ii, jj] = 1

f = radon(u, theta)

fig, ax = plt.subplots(1, 2) # plot original image and sinogram of image

ax[0].imshow(u)
ax[0].axes.get_xaxis().set_visible(False)
ax[0].axes.get_yaxis().set_visible(False)
ax[1].imshow(f, extent=(0, 2*np.pi, -1, 1), vmin=0)
ax[1].set_xlabel(r"$\theta$")
ax[1].set_ylabel(r"$s$")
ax[1].set_aspect(np.pi)

fig.tight_layout()

plt.savefig("/Users/Senne/Downloads/unitsquare.png")
```
